<h1 id="section3" style="color: #1d3557;">1-import package

```python
from bs4 import BeautifulSoup
from textblob import TextBlob
import requests
import numpy as np
import pandas as pd
import nltk
```

<h1 id="section3" style="color: #1d3557;">2- assigning the URL with variable name url

```python
url = 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
#request allow you to send HTTP request
response = requests.get(url)
soup = BeautifulSoup(response.content,'html.parser')
```

<h1 id="section3" style="color: #1d3557;">3- initialize empty lists to store the variables scarped

```python
titles = []
years = []
ratings = []
genres = []
runtimes = []
imdb_ratings = []
metascores = []
votes = []
description = []
```

<h1 id="section3" style="color: #1d3557;"> 4-storing the meaningfull required data in the variable


```python
movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})
```

<h1 id="section3" style="color: #1d3557;"> 5-calling lists  using for loop

```python
for store in movie_data:
    name = store.h3.a.text
    titles.append(name)
    
    year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
    years.append(year_of_release)
    
    my_rate = store.find('div', class_ = 'global-sprite rating-star no-rating') 
    ratings.append(my_rate)
    
    genre= store.find('span', class_='genre').text.replace('\n','' )
    genres.append(genre)
    
    runtime = store.p.find('span', class_ = 'runtime').text.replace(' min', '')
    runtimes.append(runtime)
    
    rate = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    imdb_ratings.append(rate)
    
    meta  = store.find('span', class_ = 'metascore').text.replace(' ', '') if store.find('span', class_ = 'metascore') else 'None'
    metascores.append(meta)
    #since, gross and votes have same attributes, that's why we had created a common variable and then used indexing
    value = store.find_all('span', attrs = {'name': 'nv'})
    
    vote = value[0].text
    votes.append(vote)
    
    describe = store.find_all('p', class_ = 'text-muted')
    desc = describe[1].text.replace('\n', '') if len(describe) >1 else 'None'
    description.append(desc)
   
```

<h1 id="section3" style="color: #1d3557;">6-creating a dataframe using pandas library

```python
movie_DF = pd.DataFrame({'Name of movie': titles, 'Release Date':years ,'My rate':ratings,'Genre':genres,'Movie Duration':runtimes,'Movie Rating':imdb_ratings,'Meta Score':metascores,'Number of votes':votes,'description':description}) 
```

<h1 id="section3" style="color: #1d3557;">8-Set reviews:


```python
link=input('input the film link')
page = requests.get(link)
page
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
scraped_movies = soup.findAll('div', class_='text show-more__control')
scraped_movies
movies = []
for movie in scraped_movies:
    movie = movie.get_text()
    movie = movie.strip(" ").replace('\n', "")
    movies.append(movie)
reviews=movies
```

```python
reviews
```

```python
df = pd.DataFrame()
df['Reviews of THE GOD FATHER film']=reviews
df
```

<h1 id="section3" style="color: #1d3557;">8-analyse des sentimes:

```python
list=[]

for i in range(len(reviews)):
    text=TextBlob(reviews[i])
    x=text.sentiment.polarity # x between -1 and 1
    list.append(x)
list
```

```python
sent=[]
j=[]
for j in list:
    if j>0:
        x="positive"
    elif j==0:
        x="neutre"
    else :
        x="negative"
    sent.append(x)
```

```python
df = pd.DataFrame()
df['Reviews ']=reviews
df['sentiment']=sent
df
```
